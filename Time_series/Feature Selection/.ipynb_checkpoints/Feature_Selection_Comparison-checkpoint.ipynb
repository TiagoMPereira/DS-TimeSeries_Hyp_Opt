## Comparsion between Feature Selection Methods

#### Imports

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import time

from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.feature_selection import SelectFromModel

from sklearn.linear_model import LassoCV
from sklearn.svm import SVR
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor as rf
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

#### Importando dataset

```python
from sklearn.datasets import load_boston

X,y = load_boston(return_X_y = True)
labels = load_boston().feature_names
```

#### FS Methods

```python
print("--Mutual information--")
mutual_importance = mutual_info_regression(X, y)


print("--F Regression--")
f_importance = f_regression(X, y)[0]


print("--Data Correlation--")
df = pd.DataFrame(X)
df['Target'] = y
cor = df.corr()


print("--Variance Threshold--")
vt = VarianceThreshold(threshold = 0.5)
vt.fit(X, y)
vt_importance = vt.get_support()


print("--Forward Sequential Feature Selection--")
svm = SVR()
fsfs = SequentialFeatureSelector(svm, n_features_to_select = 5)
fsfs.fit(X, y)
fsfs_importance = fsfs.get_support()

print("--Backward Sequential Feature Selection--")
bsfs = SequentialFeatureSelector(svm, n_features_to_select = 5)
bsfs.fit(X, y)
bsfs_importance = bsfs.get_support()


print("--Select From Model Lasso--")
lasso = LassoCV(cv= 5)
lasso.fit(X,y)
lasso_imp = lasso.coef_

print("--Select From Model Tree--")
tree = ExtraTreesRegressor(n_estimators = 50)
tree.fit(X,y)
tree_imp = tree.feature_importances_
tree_sfm = SelectFromModel(tree)
tree_sfm.fit(X,y)


print("--Random Forest Importance--")
rf_model = rf(n_estimators = 100)
rf_model.fit(X,y)
rf_importance = rf_model.feature_importances_
```

#### Mutual Information
Selecionando as 5 melhores features

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("Mutual Information")
plt.bar(height = mutual_importance, x = labels)
plt.plot()
```

```python
#ordenando as listas de acordo com a importância
mutual_feat, mutual_labels, mutual_indexes = zip(*sorted(zip(mutual_importance, labels, np.arange(len(labels)))))

#Resgatando as 5 features mais relevantes
mutual_feat, mutual_labels, mutual_indexes = mutual_feat[-5:], mutual_labels[-5:], mutual_indexes[-5:]

#Transformando X
X_mutual = X[:, [mutual_indexes]]
X_mutual = X_mutual.reshape(X.shape[0], 5)

print("Former shape: ", X.shape)
print("Current shape: ", X_mutual.shape)
```

#### F Regression

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("F Regression")
plt.bar(height = f_importance, x = labels)
plt.plot()
```

```python
#ordenando as listas de acordo com a importância
f_feat, f_labels, f_indexes = zip(*sorted(zip(f_importance, labels, np.arange(len(labels)))))

#Como há uma grande discrepância entre os valores de importância, serão selecionados apenas os dois melhores
f_feat, f_labels, f_indexes = f_feat[-2:], f_labels[-2:], f_indexes[-2:]

#Transformando X
X_f = X[:, [f_indexes]]
X_f = X_f.reshape(X.shape[0], 2)

print("Former shape: ", X.shape)
print("Current shape: ", X_f.shape)
```

#### Correlation Coeficient

```python
plt.figure(figsize=(12,8))
sns.heatmap(cor, annot = True)
```

```python
#Armazenando as 5 melhores features (aquelas que apresentam maior 
#correlação, em módulo, com a target)

#Maiores correlações -> 12, 5, 10, 2, 9

X_cor = X[:,[12, 5, 10, 2, 9]]

print("Former shape: ", X.shape)
print("Current shape: ", X_cor.shape)
```

#### Variance Threshold

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("Variance Threshold 0.5 (True - False)")
plt.bar(height = vt_importance, x = labels)
plt.plot()
```

```python
#Armazenando as features que passaram no teste de variância

vt_feat = [n for n,f in enumerate(vt_importance) if f]

#Transformando X
X_vt = X[:,[vt_feat]]
X_vt = X_vt.reshape(X.shape[0], X_vt.shape[2])

print("Former shape: ", X.shape)
print("Current shape: ", X_vt.shape)
```

#### Forward and Backward Sequential Feature Selection

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("Forward Sequential Feature Selection")
plt.bar(height = fsfs_importance, x = labels)
plt.plot()
```

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("Backward Sequential Feature Selection")
plt.bar(height = bsfs_importance, x = labels)
plt.plot()
```

```python
#Transformando X
X_fsfs = fsfs.transform(X)
X_bsfs = bsfs.transform(X)

print("Former shape: ", X.shape)
print("Current shape (Forward): ", X_fsfs.shape)
print("Current shape (Backward): ", X_bsfs.shape)
```

#### Select From Lasso

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("Lasso Importance")
plt.bar(height = lasso_imp, x = labels)
plt.plot()
```

```python
# 4 melhores features
threshold = np.sort(np.abs(lasso_imp))[-5]+0.01
lasso_sfm = SelectFromModel(lasso, threshold = threshold)
lasso_sfm.fit(X,y)

X_lasso = lasso_sfm.transform(X)

print("Former shape: ", X.shape)
print("Current shape: ", X_lasso.shape)
```

#### Select From Tree

```python
#Plotando ganho de informação
plt.figure(figsize=(12,8))
plt.title("Tree Importance")
plt.bar(height = tree_imp, x = labels)
plt.plot()
```

```python
#Transformando X
X_tree = tree_sfm.transform(X)

print("Former shape: ", X.shape)
print("Current shape: ", X_tree.shape)
```

#### Random Forest Importance

```python
plt.figure(figsize=(12,8))
plt.title("Random Forest Importance")
plt.bar(height = rf_importance, x = labels)
plt.show()
```

```python
# Apenas 2 features são mais relevantes

#ordenando as listas de acordo com a importância
rf_feat, rf_labels, rf_indexes = zip(*sorted(zip(rf_importance, labels, np.arange(len(labels)))))

#Selecionando as 2 feat mais relevantes
rf_feat, rf_labels, rf_indexes = rf_feat[-2:], rf_labels[-2:], rf_indexes[-2:]

#Transformando X
X_rf = X[:, [rf_indexes]]
X_rf = X_rf.reshape(X.shape[0], 2)

print("Former shape: ", X.shape)
print("Current shape: ", X_rf.shape)
```

## Testando eficácia em modelos de ML

Serão usados os modelos:
* SVM
* Decision Tree
* Random Forest

```python
svr_score = []
svr_time = []
dt_score = []
dt_time = []
rf_score = []
rf_time = []
total_score = []
total_time = []
names = ['Default','Mutual Information', 'F Regressor','Correlation',
        'Variance','Forward SFS','Backward SFS', 'Lasso','Tree','RF']

X_data = [X, X_mutual, X_f, X_cor, X_vt, X_fsfs, X_bsfs, X_lasso, X_tree, X_rf]
```

```python
for i,x in enumerate(X_data):
    
    print(names[i])
    
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=10)
    
    print("--SVR--")
    ini_svr = time.time()
    svr_model = SVR(kernel='rbf', C=1, epsilon = 1e-3)
    svr_model.fit(X_train, y_train)
    
    svr_time.append(time.time() - ini_svr)
    aux_svr = svr_model.score(X_test, y_test)
    svr_score.append(aux_svr)
    print("OK")
    
    print("--Decision Tree--")
    ini_dt = time.time()
    dt_model = DecisionTreeRegressor()
    dt_model.fit(X_train, y_train)
    
    dt_time.append(time.time() - ini_dt)
    aux_dt = dt_model.score(X_test, y_test)
    dt_score.append(aux_dt)
    print("OK")
    
    print("--Random Foreset--")
    ini_rf = time.time()
    rf_model = rf(n_estimators = 100)
    rf_model.fit(X_train, y_train)
    
    rf_time.append(time.time() - ini_rf)
    aux_rf = rf_model.score(X_test, y_test)
    rf_score.append(aux_rf)
    print("OK")
    
    total_time.append(time.time() - ini_svr)
    total_score.append((aux_svr+aux_dt+aux_rf)/3.0)
```

### Plotando Resultados

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].bar(height = svr_score, x = names)
axs[0, 0].title("SVR")
axs[0, 0].set_ylim([0, 1])
```
